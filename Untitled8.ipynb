```python


import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

```

```python
# -------------------------------------------------------------------------
# 1. Load dataset (try sklearn first; fallback to CSV in working dir)
# -------------------------------------------------------------------------
boston_df = pd.read_csv('/content/boston_housing.csv')
try:
    # sklearn has deprecated load_boston in newer versions, but this works in many environments
    from sklearn.datasets import load_boston
    data = load_boston()
    boston_df = pd.DataFrame(data.data, columns=data.feature_names)
    boston_df['MEDV'] = data.target
    print("Loaded Boston dataset via sklearn.load_boston()")
except Exception:
    # Look for a local CSV
    for root, dirs, files in os.walk('.'):
        for f in files:
            if f.lower().endswith('.csv'):
                try:
                    tmp = pd.read_csv(os.path.join(root, f))
                    # heuristic: check for MEDV or medv or RM, NOX etc
                    cols = [c.upper() for c in tmp.columns]
                    if ('MEDV' in cols) or ('NOX' in cols and 'INDUS' in cols):
                        boston_df = tmp.copy()
                        # align medv column name
                        if 'medv' in boston_df.columns:
                            boston_df.rename(columns={'medv': 'MEDV'}, inplace=True)
                        print(f"Loaded dataset from {os.path.join(root,f)}")
                        break
                except Exception:
                    continue
        if boston_df is not None:
            break

if boston_df is None:
    raise FileNotFoundError("Could not load Boston dataset. Install sklearn with old datasets or upload a CSV containing the dataset.")

# Ensure required columns exist and proper types
required_cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','LSTAT','MEDV']
missing = [c for c in required_cols if c not in boston_df.columns]
if missing:
    raise ValueError(f"Missing required columns: {missing}")

# Ensure CHAS integer (0/1)
boston_df['CHAS'] = boston_df['CHAS'].astype(int)

# Quick head and basic stats
print("\nDataset shape:", boston_df.shape)
print(boston_df[required_cols].head())

# -------------------------------------------------------------------------
# 2. Boxplot for MEDV (Median value of owner-occupied homes)
# -------------------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.boxplot(boston_df['MEDV'].dropna(), vert=True, labels=['MEDV'])
plt.title('Boxplot of Median Value of Owner-Occupied Homes (MEDV) — in $1000s')
plt.ylabel('Median value (1000s USD)')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()

print("\nMEDV basic statistics:\n", boston_df['MEDV'].describe().to_frame().T)

# -------------------------------------------------------------------------
# 3. Bar plot for the CHAS variable (counts)
# -------------------------------------------------------------------------
chas_counts = boston_df['CHAS'].value_counts().sort_index()
plt.figure(figsize=(6,4))
plt.bar(['0 - Not bounding Charles River','1 - Bounds Charles River'], chas_counts.values)
plt.title('Counts of Tracts by CHAS (Bounds Charles River?)')
plt.xlabel('CHAS (0 = No, 1 = Yes)')
plt.ylabel('Number of tracts')
for i, v in enumerate(chas_counts.values):
    plt.text(i, v + 1, str(v), ha='center')
plt.show()

print("\nCHAS counts:\n", chas_counts.to_string())

# -------------------------------------------------------------------------
# 4. Boxplot MEDV vs AGE (discretize AGE into three groups)
#    User requested groups:
#      - 35 years and younger
#      - between 35 and 70 years
#      - 70 years and older
#    (AGE is proportion built prior to 1940; values in [0,100])
# -------------------------------------------------------------------------
# Construct bins so that boundaries follow the user's definition.
bins = [-0.1, 35.0, 70.0, 100.0]   # include 35 in first group per "35 years and younger"
labels = ['<=35 years', '35-70 years', '>=70 years']
boston_df['AGE_group'] = pd.cut(boston_df['AGE'], bins=bins, labels=labels, include_lowest=True, right=True)

# Boxplot
plt.figure(figsize=(9,6))
grouped_data = [boston_df.loc[boston_df['AGE_group'] == lab, 'MEDV'].dropna() for lab in labels]
plt.boxplot(grouped_data, labels=labels)
plt.title('MEDV by AGE group (proportion of owner-occupied units built prior to 1940)')
plt.xlabel('AGE group (percentage built before 1940)')
plt.ylabel('Median value (1000s USD)')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()

# Print group sizes and summary stats
age_group_stats = boston_df.groupby('AGE_group')['MEDV'].agg(['count','mean','median','std']).reset_index()
print("\nMEDV statistics by AGE group:\n", age_group_stats.to_string(index=False))

# -------------------------------------------------------------------------
# 5. Scatter plot: NOX vs INDUS (relationship between Nitric oxide concentration and
#    proportion of non-retail business acres per town)
# -------------------------------------------------------------------------
plt.figure(figsize=(8,6))
plt.scatter(boston_df['INDUS'], boston_df['NOX'], alpha=0.7)
# linear fit
m, c = np.polyfit(boston_df['INDUS'], boston_df['NOX'], 1)
xvals = np.linspace(boston_df['INDUS'].min(), boston_df['INDUS'].max(), 100)
plt.plot(xvals, m*xvals + c, linestyle='--', linewidth=1)
plt.title('NOX (Nitric oxide conc.) vs INDUS (Non-retail business acres proportion)')
plt.xlabel('INDUS (proportion of non-retail business acres per town)')
plt.ylabel('NOX (nitric oxides concentration, parts per 10 million)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()

# Pearson correlation:
pearson_r, pearson_p = stats.pearsonr(boston_df['INDUS'], boston_df['NOX'])
print(f"\nPearson correlation NOX vs INDUS: r = {pearson_r:.4f}, p-value = {pearson_p:.4e}")

# -------------------------------------------------------------------------
# 6. Histogram for PTRATIO (pupil-to-teacher ratio)
# -------------------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.hist(boston_df['PTRATIO'].dropna(), bins=12, edgecolor='k')
plt.title('Histogram of Pupil-Teacher Ratio by Town (PTRATIO)')
plt.xlabel('Pupil-Teacher Ratio')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()

print("\nPTRATIO summary:\n", boston_df['PTRATIO'].describe().to_frame().T)

# -------------------------------------------------------------------------
# 7. Statistical tests
#    Use alpha = 0.05
#    (a) T-test for MEDV by CHAS (independent samples; Welch's t-test)
#    (b) ANOVA for MEDV across AGE groups
#    (c) Pearson correlation NOX vs INDUS
#    (d) Simple regression: MEDV ~ DIS (report slope, p-value, R^2)
# -------------------------------------------------------------------------
alpha = 0.05
print("\n\n--- STATISTICAL TESTS (alpha = 0.05) ---")

# (a) T-test: Is median/mean MEDV different for tracts bounding Charles River (CHAS=1) vs not (CHAS=0)?
medv_chas1 = boston_df[boston_df['CHAS'] == 1]['MEDV'].dropna()
medv_chas0 = boston_df[boston_df['CHAS'] == 0]['MEDV'].dropna()

print("\n(a) Independent t-test (Welch) — MEDV by CHAS")
print("H0: mean(MEDV | CHAS=1) = mean(MEDV | CHAS=0)")
print("H1: means differ")
t_stat, p_value = stats.ttest_ind(medv_chas1, medv_chas0, equal_var=False, nan_policy='omit')
print(f" n(CHAS=1) = {medv_chas1.size}, n(CHAS=0) = {medv_chas0.size}")
print(f" median(MEDV) CHAS=1: {medv_chas1.median():.3f}, CHAS=0: {medv_chas0.median():.3f}")
print(f" t-statistic = {t_stat:.4f}, p-value = {p_value:.4e}")
if p_value < alpha:
    print(" Conclusion: Reject H0 — there is a statistically significant difference in MEDV between CHAS groups.")
else:
    print(" Conclusion: Fail to reject H0 — no evidence of a significant difference at alpha=0.05.")

# (b) ANOVA: MEDV across AGE groups
print("\n(b) One-way ANOVA — MEDV across AGE groups (<=35, 35-70, >=70)")
print("H0: All group means are equal")
print("H1: At least one group mean differs")
groups = [boston_df[boston_df['AGE_group'] == label]['MEDV'].dropna() for label in labels]
f_stat, p_val_anova = stats.f_oneway(*groups)
print(f" F-statistic = {f_stat:.4f}, p-value = {p_val_anova:.4e}")
if p_val_anova < alpha:
    print(" Conclusion: Reject H0 — at least one AGE group has a different mean MEDV.")
else:
    print(" Conclusion: Fail to reject H0 — no evidence that group means differ at alpha=0.05.")

# If ANOVA significant, show pairwise comparisons (Tukey HSD)
if p_val_anova < alpha:
    try:
        from statsmodels.stats.multicomp import pairwise_tukeyhsd
        # Prepare data for Tukey
        tukey = pairwise_tukeyhsd(endog=boston_df['MEDV'], groups=boston_df['AGE_group'], alpha=0.05)
        print("\nTukey HSD pairwise comparisons (AGE groups):")
        print(tukey.summary())
    except Exception as e:
        print("Could not run Tukey HSD (missing dependency).", e)

# (c) Pearson correlation NOX vs INDUS
print("\n(c) Pearson correlation — NOX vs INDUS")
print("H0: rho = 0 (no linear correlation)")
print("H1: rho != 0")
print(f" r = {pearson_r:.4f}, p-value = {pearson_p:.4e}")
if pearson_p < alpha:
    print(" Conclusion: Reject H0 — significant linear relationship between NOX and INDUS.")
else:
    print(" Conclusion: Fail to reject H0 — no evidence of linear relationship.")

# (d) Regression: MEDV ~ DIS (impact of weighted distance to five Boston employment centers)
print("\n(d) Linear regression: MEDV ~ DIS")
print("Model: MEDV = intercept + slope * DIS + error")
X = boston_df[['DIS']].values
y = boston_df['MEDV'].values
lr = LinearRegression().fit(X, y)
slope = lr.coef_[0]
intercept = lr.intercept_
r2 = lr.score(X, y)

# Use statsmodels OLS to get p-value and t-stat
X_sm = sm.add_constant(X)
ols_res = sm.OLS(y, X_sm).fit()
slope_t = ols_res.tvalues[1]
slope_p = ols_res.pvalues[1]
conf_int = ols_res.conf_int()[1].tolist()  # slope CI

print(f" slope = {slope:.4f} (change in MEDV per unit increase in DIS)")
print(f" intercept = {intercept:.4f}")
print(f" R-squared = {r2:.4f}")
print(f" t-statistic (slope) = {slope_t:.4f}, p-value = {slope_p:.4e}")
print(f" 95% CI for slope: [{conf_int[0]:.4f}, {conf_int[1]:.4f}]")
if slope_p < alpha:
    print(" Conclusion: Slope is statistically significant — DIS has a measurable (linear) association with MEDV.")
else:
    print(" Conclusion: Slope not significant at alpha=0.05 — no evidence DIS predicts MEDV.")

# Show regression scatter with fitted line
plt.figure(figsize=(8,6))
plt.scatter(boston_df['DIS'], boston_df['MEDV'], alpha=0.6)
x_vals = np.linspace(boston_df['DIS'].min(), boston_df['DIS'].max(), 100)
plt.plot(x_vals, intercept + slope * x_vals, linestyle='--')
plt.title('Scatter plot: MEDV vs DIS (with fitted regression line)')
plt.xlabel('DIS (weighted distance to five Boston employment centres)')
plt.ylabel('MEDV (Median value of owner-occupied homes, $1000s)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()

# -------------------------------------------------------------------------
# End of analysis
# -------------------------------------------------------------------------
print("\nAnalysis complete. Save figures manually if needed (e.g., plt.savefig()).")
```
