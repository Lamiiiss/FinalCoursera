```python


import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

```

```python
# --------------------------------------------------------------
# 1. Load dataset
# --------------------------------------------------------------
try:
    from sklearn.datasets import fetch_openml
    boston = fetch_openml(name='boston', version=1, as_frame=True)
    df = boston.frame
    df['MEDV'] = boston.target
    # Convert CHAS to numeric
    df['CHAS'] = pd.to_numeric(df['CHAS'])
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Please ensure the dataset is available or check the path.")
    # You might want to handle this error more gracefully, e.g., exit or load from a different source
    exit()


print("First five rows of dataset:")
display(df.head())

# --------------------------------------------------------------
# 2. Boxplot for MEDV
# --------------------------------------------------------------
import seaborn as sns
plt.figure(figsize=(6,4))
sns.boxplot(y=df['MEDV'])
plt.title("Boxplot of MEDV (Median Value of Homes)")
plt.ylabel("MEDV ($1000's)")
plt.show()
print("Explanation: The boxplot shows the distribution of home values. "
      "We see outliers above 50 (capped values). Most homes fall between $15k and $25k.\n")

# --------------------------------------------------------------
# 3. Barplot for CHAS
# --------------------------------------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x='CHAS', data=df)
plt.title("Barplot of CHAS (Charles River variable)")
plt.xlabel("CHAS (1 = tract bounds river, 0 = otherwise)")
plt.ylabel("Count")
plt.show()
print("Explanation: Most tracts are not near the Charles River (CHAS=0). "
      "Only a small number of tracts have CHAS=1.\n")

# --------------------------------------------------------------
# 4. Boxplot MEDV vs AGE
# --------------------------------------------------------------
def age_group(x):
    if x < 35:
        return "Young (<35)"
    elif 35 <= x < 70:
        return "Middle (35-70)"
    else:
        return "Old (>=70)"

df['AGE_GROUP'] = df['AGE'].apply(age_group)

plt.figure(figsize=(6,4))
sns.boxplot(x='AGE_GROUP', y='MEDV', data=df)
plt.title("Boxplot of MEDV vs AGE group")
plt.xlabel("House Age Group")
plt.ylabel("MEDV ($1000's)")
plt.show()
print("Explanation: Older houses (>=70 years) tend to have lower median values. "
      "Newer homes (<35 years) tend to have higher values.\n")

# --------------------------------------------------------------
# 5. Scatter Plot NOX vs INDUS
# --------------------------------------------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x='INDUS', y='NOX', data=df)
plt.title("Scatter plot of NOX vs INDUS")
plt.xlabel("INDUS (% non-retail business acres)")
plt.ylabel("NOX (nitric oxide concentration)")
plt.show()
print("Explanation: Positive relationship observed â€” higher industrial activity "
      "is associated with higher NOX pollution.\n")

# --------------------------------------------------------------
# 6. Histogram of PTRATIO
# --------------------------------------------------------------
plt.figure(figsize=(6,4))
sns.histplot(df['PTRATIO'], bins=15, kde=True)
plt.title("Histogram of PTRATIO (Pupil-Teacher Ratio)")
plt.xlabel("PTRATIO")
plt.ylabel("Frequency")
plt.show()
print("Explanation: Most towns have pupil-teacher ratios between 16 and 20.\n")

# --------------------------------------------------------------
# 7. Hypothesis Testing
# --------------------------------------------------------------

# --- T-test (CHAS vs MEDV)
print("T-test: Difference in MEDV by Charles River proximity")
print("H0: No difference in MEDV between tracts near river and not near river")
print("H1: Significant difference in MEDV between tracts near river and not near river")
chas1 = df[df['CHAS']==1]['MEDV']
chas0 = df[df['CHAS']==0]['MEDV']
t_stat, p_val = stats.ttest_ind(chas1, chas0)
print(f"T-stat = {t_stat:.4f}, p-value = {p_val:.4e}")
if p_val < 0.05:
    print("Conclusion: Reject H0. There IS a significant difference in home values.\n")
else:
    print("Conclusion: Fail to reject H0. No significant difference found.\n")

# --- ANOVA (AGE groups vs MEDV)
print("ANOVA: Difference in MEDV across AGE groups")
print("H0: No difference in MEDV across age groups")
print("H1: At least one group has a significantly different MEDV")
anova_res = stats.f_oneway(
    df[df['AGE_GROUP']=="Young (<35)"]['MEDV'],
    df[df['AGE_GROUP']=="Middle (35-70)"]['MEDV'],
    df[df['AGE_GROUP']=="Old (>=70)"]['MEDV']
)
print(f"F-stat = {anova_res.statistic:.4f}, p-value = {anova_res.pvalue:.4e}")
if anova_res.pvalue < 0.05:
    print("Conclusion: Reject H0. There IS a significant difference across groups.\n")
else:
    print("Conclusion: Fail to reject H0. No significant difference found.\n")

# --- Pearson correlation (NOX vs INDUS)
print("Correlation Test: NOX vs INDUS")
print("H0: No relationship between NOX and INDUS")
print("H1: Significant relationship between NOX and INDUS")
corr, p_corr = stats.pearsonr(df['NOX'], df['INDUS'])
print(f"Correlation coefficient = {corr:.4f}, p-value = {p_corr:.4e}")
if p_corr < 0.05:
    print("Conclusion: Reject H0. Significant positive correlation exists.\n")
else:
    print("Conclusion: Fail to reject H0. No significant relationship.\n")

# --- Regression (MEDV vs DIS)
print("Regression: Impact of DIS (distance to employment centers) on MEDV")
X = sm.add_constant(df[['DIS']])
y = df['MEDV']
model = sm.OLS(y, X).fit()
print(model.summary())
coef = model.params['DIS']
print(f"Coefficient of DIS = {coef:.4f}")
if coef > 0:
    print("Interpretation: Greater distance is associated with slightly higher home values.\n")
else:
    print("Interpretation: Greater distance is associated with lower home values.\n")
```
